```python
args = Namespace()

args.seed = 1
args.batch_size = 10
args.epochs = 10000
args.lr = 1e-3

args.n_dim = 30         # Size of the matrix (n x n)
args.rank = 5           # Rank of X*
args.m_samples = 5 * args.n_dim * args.rank   # Number of samples
args.m_train = args.n_dim * args.rank if args.n_dim > 10 else 25      # Number of training samples

args.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Reproducibility
torch.manual_seed(args.seed)
np.random.seed(args.seed)
random.seed(args.seed)
```

```python
# Generate low-rank matrix X_star = U U^T
U_true = torch.randn(args.n_dim, args.rank, device=args.device)
X_star = U_true @ U_true.T  # [n, n]
X_star /= torch.norm(X_star, p='fro')  # Normalize to unit Frobenius norm

# Generate random sensing matrices A_i and labels y_i = Tr(A_i^T X_star)
A = torch.randn(args.m_samples, args.n_dim, args.n_dim, device=args.device)   # [m, n, n]
y = torch.einsum('mij,ij->m', A, X_star)  # [m]


# Split into train/test
A_train, y_train = A[:args.m_train], y[:args.m_train]
A_test, y_test = A[args.m_train:], y[args.m_train:]

print(y_train.mean(), y_train.std())
# y_train += torch.randn_like(y_train) * 0.2  # Add noise to training labels
```

```python
class MatrixSensingModel(nn.Module):
    def __init__(self, args):
        super().__init__()
        U_init = torch.eye(args.n_dim, args.n_dim, device=args.device)
        U_init /= torch.norm(U_init, p='fro')  # Normalize to unit Frobenius norm
        self.U = nn.Parameter(U_init)  # Initialize U as a learnable parameter

    def forward(self):
        return self.U @ self.U.T  # X = UU^T


def compute_loss(model, A, y):
    X = model()
    AX = torch.einsum('mij,ij->m', A, X)
    residual = AX - y
    return 0.5 * torch.mean(residual ** 2)


def train(model, optimizer, A_train, y_train, A_test, y_test, args, inject_noise=False):
    train_losses = [compute_loss(model, A_train, y_train).item()]
    test_losses = [compute_loss(model, A_test, y_test).item()]

    tot_t = 1

    label_noise = torch.randn(args.m_train, device=args.device) * args.sigma if inject_noise else 0
    if hasattr(args, 'tau') and args.tau > 0:
        vt = label_noise / np.sqrt(2-args.lr/args.tau)  # Scale noise if tau is specified

    n_batch = args.m_train // args.batch_size
    tot_epochs = args.tot_iter // n_batch
    for _ in range(tot_epochs):
        # Shuffle indices for each epoch
        if n_batch == 1:
            indices = torch.arange(args.m_train, device=args.device)
        else:
            indices = torch.randperm(args.m_train, device=args.device)

        if inject_noise:
            if hasattr(args, 'tau') and args.tau > 0:
                vt_next = (1 - args.lr/args.tau) * vt + args.sigma * np.sqrt(args.lr/(args.tau**2)) * torch.randn(args.m_train, device=args.device)
                label_noise = args.tau * (vt_next - vt)
                vt = vt_next
                # label_noise = (1 - args.lr/args.tau) * label_noise + args.sigma * np.sqrt(args.lr/args.tau) * torch.randn(args.m_train, device=args.device)
            else:
                label_noise = torch.randn(args.m_train, device=args.device) * args.sigma if inject_noise else 0

        y_train_noisy = y_train + label_noise

        for i in range(0, args.m_train, args.batch_size):
            batch_idx = indices[i:i + args.batch_size]
            A_batch = A_train[batch_idx]
            y_batch = y_train_noisy[batch_idx]

            optimizer.zero_grad()
            loss = compute_loss(model, A_batch, y_batch)
            loss.backward()
            optimizer.step()

            # Logging
            if tot_t % args.log_iter == 0:  # Optional: log less frequently
                with torch.no_grad():
                    train_loss = compute_loss(model, A_train, y_train).item()
                    test_loss = compute_loss(model, A_test, y_test).item()
                    print(f"Iter {tot_t}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}")
                    train_losses.append(train_loss)
                    test_losses.append(test_loss)
                
                # Optional: update the log_iter dynamically
                # if tot_t > 0 and tot_t % (args.log_iter * 10) == 0:
                #     args.log_iter *= 10  # Increase log_iter to log less frequently

            tot_t += 1

    return {'train_loss': train_losses, 'test_loss': test_losses}
```

```python
def hessian_trace_rademacher(loss_fn, U_, A, y, num_samples=100, tol=1e-4):
    U = U_.clone().requires_grad_(True)  # Safe local copy

    trace_estimates = []
    trace = 0.

    for _ in range(num_samples):
        # Rademacher vector
        v = torch.randint_like(U, low=0, high=2, dtype=torch.float32)
        v = 2.0 * v - 1.0  # in {-1, +1}
        v_flat = v.view(-1)

        # Compute gradient
        loss = loss_fn(U, A, y)
        grad = torch.autograd.grad(loss, U, create_graph=True)[0]
        grad_flat = grad.view(-1)

        Hv = torch.autograd.grad(
            grad_flat, U,
            grad_outputs=v_flat,
            retain_graph=True,
            allow_unused=True
        )[0]

        if Hv is None:
            continue

        trace_estimates.append(torch.sum(v * Hv).cpu().item())
        if abs(np.mean(trace_estimates) - trace)/(abs(trace) + 1e-6) < tol:
            return trace_estimates
        else:
            trace = np.mean(trace_estimates)

    return trace_estimates
```
