```python
args = Namespace()

args.seed = 1
args.batch_size = 10
args.epochs = 10000
args.lr = 1e-3

args.n_dim = 50         # Size of the matrix (n x n)
args.rank = 5           # Rank of X*
args.m_samples = 10 * args.n_dim * args.rank   # Number of samples
args.m_train = args.n_dim * args.rank      # Number of training samples

args.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Reproducibility
torch.manual_seed(args.seed)
np.random.seed(args.seed)
random.seed(args.seed)
```

```python
# Generate low-rank matrix X_star = U U^T
U_true = torch.randn(args.n_dim, args.rank, device=args.device)
X_star = U_true @ U_true.T  # [n, n]
X_star /= torch.norm(X_star, p='fro')  # Normalize to unit Frobenius norm

# Generate random sensing matrices A_i and labels y_i = Tr(A_i^T X_star)
A = torch.randn(args.m_samples, args.n_dim, args.n_dim, device=args.device)   # [m, n, n]
y = torch.einsum('mij,ij->m', A, X_star)  # [m]


# Split into train/test
A_train, y_train = A[:args.m_train], y[:args.m_train]
A_test, y_test = A[args.m_train:], y[args.m_train:]

print(y_train.mean(), y_train.std())

# y_train += torch.randn_like(y_train) * 0.1  # Add noise to training labels
```

```python
# Loss
def loss(U, A, y):
    X = U @ U.T
    residuals = torch.einsum('mij,ij->m', A, X) - y
    return 0.5 * torch.mean(residuals ** 2)

# Gradient
def grad(U, A, y):
    # U: [n, r], A: [m, n, n], y: [m]
    X = U @ U.T                             # [n, n]
    AX = torch.einsum('mij,ij->m', A, X)   # [m], compute Tr(A_i X)
    residuals = AX - y                     # [m]
    
    # Compute batched matrix products: A_i @ U â†’ [m, n, r]
    AU = torch.einsum('mij,jk->mik', A, U)   # [m, n, r]

    # Weight each A_i @ U by 2 * residual[i]
    grad_tensor = 2 * residuals.view(-1, 1, 1) * AU  # [m, n, r]

    # Average over the batch
    return grad_tensor.mean(dim=0)  # [n, r]

# Training function
def train_gd(U_init, A_train, y_train, A_test, y_test, args, inject_noise=False):
    U = deepcopy(U_init).to(args.device)

    train_losses, test_losses = [], []

    label_noise = torch.randn(args.m_train, device=args.device) * args.sigma if inject_noise else 0
    if hasattr(args, 'tau') and args.tau > 0:
        label_noise = label_noise / np.sqrt(2-args.lr/args.tau)  # Scale noise if tau is specified

    for t in range(args.tot_iter+1):
        if inject_noise:
            if hasattr(args, 'tau') and args.tau > 0:
                label_noise = (1 - args.lr/args.tau) * label_noise + args.sigma * np.sqrt(args.lr/args.tau) * torch.randn(args.m_train, device=args.device)
            else:
                label_noise = torch.randn(args.m_train, device=args.device) * args.sigma

        y_train_noisy = y_train + label_noise  # Add noise to training labels if specified

        g = grad(U, A_train, y_train_noisy)
        U -= args.lr * g            

        if t % args.log_iter == 0:  # Optional: log less frequently
            train_losses.append(loss(U, A_train, y_train).item())
            test_losses.append(loss(U, A_test, y_test).item())
            
    return {'train_loss': train_losses, 'test_loss': test_losses}, U


# Training
def train_sgd(U_init, A_train, y_train, A_test, y_test, args, inject_noise=False):
    U = deepcopy(U_init).to(args.device)

    train_losses, test_losses = [], []

    tot_t = 1
    label_noise = torch.randn(args.m_train, device=args.device) * args.sigma if inject_noise else 0

    n_batch = len(A_train) // args.batch_size
    tot_epochs = args.tot_iter // n_batch
    for _ in range(tot_epochs):
        if inject_noise:
            if hasattr(args, 'tau') and args.tau > 0:
                label_noise = (1 - args.lr/args.tau) * label_noise + args.sigma * np.sqrt(args.lr/args.tau) * torch.randn(args.m_train, device=args.device)
            else:
                label_noise = torch.randn(args.m_train, device=args.device) * args.sigma if inject_noise else 0

        y_train_noisy = y_train + label_noise

        indices = torch.randperm(len(A_train), device=args.device)
        for i in range(0, len(A_train), args.batch_size):
            batch_idx = indices[i:i + args.batch_size]
            A_batch = A_train[batch_idx]
            y_batch = y_train_noisy[batch_idx]

            g = grad(U, A_batch, y_batch)
            U -= args.lr * g            

            if tot_t % args.log_iter == 0:  # Optional: log less frequently
                train_losses.append(loss(U, A_train, y_train).item())
                test_losses.append(loss(U, A_test, y_test).item())

            tot_t += 1
            
    return {'train_loss': train_losses, 'test_loss': test_losses}, U
```

```python
args.log_iter = 1000  # Logging interval
args.tot_iter = args.log_iter * 10**2  # Number of epochs

args.sigma = 0.1 # Standard deviation of noise
args.lr = 1e-2
args.batch_size = args.m_train // 10 
args.tau = 0

# Initialize and run
U0 = torch.eye(args.n_dim, device=args.device)
U0 /= torch.norm(U0, p='fro')

losses_clean, U_clean = train_sgd(U0, A_train, y_train, A_test, y_test, args, inject_noise=False)
losses_noisy, U_noisy = train_sgd(U0, A_train, y_train, A_test, y_test, args, inject_noise=True)

args.tau = 5 * args.lr
losses_noisy_tau, U_noisy_tau = train_sgd(U0, A_train, y_train, A_test, y_test, args, inject_noise=True)
```

```python
def hessian_trace_rademacher(loss_fn, U_, A, y, num_samples=100, tol=1e-4):
    U = U_.clone().requires_grad_(True)  # Safe local copy

    trace_estimates = []
    trace = 0.

    for _ in range(num_samples):
        # Rademacher vector
        v = torch.randint_like(U, low=0, high=2, dtype=torch.float32)
        v = 2.0 * v - 1.0  # in {-1, +1}
        v_flat = v.view(-1)

        # Compute gradient
        loss = loss_fn(U, A, y)
        grad = torch.autograd.grad(loss, U, create_graph=True)[0]
        grad_flat = grad.view(-1)

        Hv = torch.autograd.grad(
            grad_flat, U,
            grad_outputs=v_flat,
            retain_graph=True,
            allow_unused=True
        )[0]

        if Hv is None:
            continue

        trace_estimates.append(torch.sum(v * Hv).cpu().item())
        if abs(np.mean(trace_estimates) - trace)/(abs(trace) + 1e-6) < tol:
            return trace_estimates
        else:
            trace = np.mean(trace_estimates)

    return trace_estimates
```
